# Building AI Agents with Vector Databases: Contract Analysis Agent
## Notebook 2: Building the Contract Analysis Agent

## Setup and connection

```python
import os
from dotenv import load_dotenv
import weaviate
from weaviate.classes.init import Auth

load_dotenv()

WEAVIATE_URL = os.getenv("WEAVIATE_URL")
WEAVIATE_KEY = os.getenv("WEAVIATE_KEY")

client = weaviate.connect_to_weaviate_cloud(
    cluster_url=WEAVIATE_URL,
    auth_credentials=Auth.api_key(WEAVIATE_KEY),
)

print("Client ready:", client.is_ready())

# Get our contracts collection
contracts = client.collections.get("FinancialContract")
print(f"Contracts available: {len(contracts)}")
```

## Installing the Query Agent

The Weaviate Query Agent is a pre-built agentic service that can answer natural language questions about your data.

```python
from weaviate.agents.query import QueryAgent

# Create our contract analysis agent
agent = QueryAgent(
    client=client,
    collection_names=["FinancialContract"],
    model="gpt-4o-mini",
    system_message="You are a legal contract analysis expert. Help users understand contracts, identify risks, and suggest improvements. Always provide specific, actionable advice."
)

print("Contract Analysis Agent created!")
```

## Agent Task 1: Risk Assessment

```python
print("=== Contract Risk Analysis ===")
risk_query = "What are the top 3 risks in employment contracts that employers should watch out for? Provide specific examples."

response = agent.query(risk_query)
print("Agent Response:")
print(response)
```

```python
print("=== Payment Risk Analysis ===")
payment_risk_query = "Find contracts with payment terms and analyze what payment-related risks the paying party might face."

response = agent.query(payment_risk_query)
print("Agent Response:")
print(response)
```

## Agent Task 2: Contract Comparison

```python
print("=== Contract Comparison ===")
comparison_query = "Find 2 similar service agreements and compare their termination clauses. Which one is more favorable for the service provider?"

response = agent.query(comparison_query)
print("Agent Response:")
print(response)
```

## Agent Task 3: Contract Drafting Suggestions

```python
print("=== Contract Improvement Suggestions ===")
improvement_query = "Based on the contracts in the database, what are 5 essential clauses that every service agreement should include to protect both parties?"

response = agent.query(improvement_query)
print("Agent Response:")
print(response)
```

## Multi-Step Agent Workflows

```python
def comprehensive_contract_analysis(contract_type):
    print(f"=== Comprehensive Analysis: {contract_type} ===")
    
    # Step 1: Find relevant contracts
    search_query = f"Find {contract_type} contracts in the database"
    print("Step 1: Finding contracts...")
    contracts_found = agent.query(search_query)
    print("Contracts found:", contracts_found[:200] + "..." if len(contracts_found) > 200 else contracts_found)
    
    # Step 2: Risk assessment
    risk_query = f"Analyze the main risks in {contract_type} contracts and rank them by severity"
    print("Step 2: Risk assessment...")
    risk_analysis = agent.query(risk_query)
    print("Risk analysis:", risk_analysis[:300] + "..." if len(risk_analysis) > 300 else risk_analysis)
    
    return {"contracts_found": contracts_found, "risk_analysis": risk_analysis}

# Run comprehensive analysis
employment_analysis = comprehensive_contract_analysis("employment")
```

## Business Scenarios

```python
print("=== Scenario: Vendor Selection ===")
vendor_query = "A company needs to choose between vendor contracts. What criteria should they use to evaluate and compare vendor agreements?"

response = agent.query(vendor_query)
print("Vendor Selection Guidance:")
print(response)
```

```python
print("=== Scenario: Red Flag Detection ===")
red_flag_query = "What are the biggest red flags in contracts that businesses should never agree to?"

response = agent.query(red_flag_query)
print("Contract Red Flags:")
print(response)
```
